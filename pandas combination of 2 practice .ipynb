```python
import pandas as pd 
df = pd.DataFrame({'name':['a','a','a'],'items':['b','c','d']})
df_a = df.copy()
df_b = df.copy()


df_a['constant'] = 1
df_b['constant'] = 1


merge_df = df_a.merge(df_b,on=['constant','name'], suffixes=['_a','_b'])


merge_df['total_items'] = merge_df.apply(lambda row: \
	'&'.join(list(set([row['items_a'],row['items_b']] ))) ,axis=1)

merge_df.drop_duplicates('total_items', inplace = True)
print(merge_df)
```
